```python
### Load Data
# we can convert this into one  function for finance

import yfinance as yf
def fix_col(df):
    col = df.columns
    outer= [cols[0] for cols in col]
    df.columns = outer
    return df
def tweak():
    data = yf.download('SPY AAPL',start='2010-01-01',end='2023-06-27')
    return(data.iloc[:,::2].pipe(fix_col))
tweak()
```

```python
### Line Plots

(data
 .iloc[:,:-2:2]
 .pipe(fix_col)
 .plot())
```

```python
(data
 .iloc[:,::2]
 .pipe(fix_col)
 .Close
 .plot())
```

```python
#Resampling : taking data from one granulatiy to other granuality like from daily to weekly monthly of quarterly
(data
.iloc[:,::2]
.pipe(fix_col)
.resample('M')
.Close
.mean()
.plot() 
)
```

```python
(data
.iloc[:,::2]
.pipe(fix_col)
.resample('Q')
.Close
.mean()
.plot() 
)
```

```python
#Candle stick plot
fig,ax = plt.subplots(figsize=(10,5))
def candlestick(df,ax):
    ax.vlines(x=df.index, ymin=df.Low, ymax=df.High, color='r', linewidth=1)
    red = df.query('Open > Close')
    ax.vlines(x=red.index,ymin=red.Close, ymax=red.Open, color='r', linewidth=3)
    green = df.query('Open <= Close')
    ax.vlines(x=green.index,ymin=green.Open, ymax=green.Close, color='g', linewidth=3)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return df
(data
.iloc[:,::2]
.pipe(fix_col)
.resample('d')
.agg({'Close':'last', 'Open':'first',  'High':'max', 'Low':'min'})
.loc['jan 2018':'March 2018']
.pipe(candlestick,ax) 
)
```

```python
#Candle stick plot
fig,ax = plt.subplots(figsize=(10,5))
def candlestick(df,ax):
    ax.vlines(x=df.index, ymin=df.Low, ymax=df.High, color='r', linewidth=1)
    red = df.query('Open > Close')
    ax.vlines(x=red.index,ymin=red.Close, ymax=red.Open, color='r', linewidth=3)
    green = df.query('Open <= Close')
    ax.vlines(x=green.index,ymin=green.Open, ymax=green.Close, color='g', linewidth=3)
    ax.xaxis.set_major_locator(dates.MonthLocator())
    ax.xaxis.set_major_formatter(dates.DateFormatter('%b-%y'))
    ax.xaxis.set_minor_locator(dates.DayLocator())
    return df
(data
.iloc[:,::2]
.pipe(fix_col)
.resample('d')
.agg({'Close':'last', 'Open':'first',  'High':'max', 'Low':'min'})
.loc['jan 2018':'March 2018']
.pipe(candlestick,ax) 
)
```

### Volality

volatility refers to the degree of variation or fluctuation in the price or value of a financial instrument, such as stocks, bonds, or commodities, over a specific period. It is a measure of the uncertainty or risk associated with an investment.

Volatility is commonly used to assess the potential price movements and riskiness of an asset. Higher volatility indicates larger price swings and greater uncertainty, while lower volatility suggests more stable and predictable price movements.

There are different methods to measure volatility, but two commonly used measures are:

Standard Deviation: The standard deviation is a statistical measure that quantifies the dispersion of a 
dataset from its mean. In finance, it is often used to calculate historical volatility. 
By analyzing the historical price data, the standard deviation provides an estimate of 
how much the prices have deviated from the average in the past. Higher standard deviation 
indicates higher volatility.

Volatility Index (VIX): The VIX is a popular index that measures the implied volatility of the options market.
It reflects the market's expectations of future volatility. The VIX is often referred to as the "fear index" as it tends to rise during times of market stress and uncertainty.
Volatility plays a crucial role in various aspects of finance, such as risk management, option pricing, portfolio construction, and trading strategies. It helps investors and traders assess the potential risks and rewards associated with different investments and make informed decisions.

It's important to note that volatility is not the same as the direction of price movements. 
A highly volatile asset can have significant price swings in both positive and negative directions, 
while a less volatile asset may have relatively smaller and more predictable price movements.

By understanding and analyzing volatility, market participants can better evaluate the potential 
risks and returns of their investments and adapt their strategies accordingly.

### Rolling

Concept of "rolling" refers to a technique of calculating metrics or performing calculations repeatedly on a moving or rolling window of data. Rolling calculations are commonly used to analyze time-series data, such as stock prices, economic indicators, or financial ratios, in order to observe trends or patterns over a specified period.

The rolling window typically consists of a fixed number of consecutive observations or a defined time period. As new data becomes available, the window moves forward in time, and the calculations are updated using the latest observations while dropping the oldest ones.

With rolling  we can have rolling mean ( Moving averages), Rolling standard deviation (Volality), rolling sum, rolling corelations

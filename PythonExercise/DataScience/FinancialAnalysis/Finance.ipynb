### Volality

volatility refers to the degree of variation or fluctuation in the price or value of a financial instrument, such as stocks, bonds, or commodities, over a specific period. It is a measure of the uncertainty or risk associated with an investment.

Volatility is commonly used to assess the potential price movements and riskiness of an asset. Higher volatility indicates larger price swings and greater uncertainty, while lower volatility suggests more stable and predictable price movements.

There are different methods to measure volatility, but two commonly used measures are:

Standard Deviation: The standard deviation is a statistical measure that quantifies the dispersion of a 
dataset from its mean. In finance, it is often used to calculate historical volatility. 
By analyzing the historical price data, the standard deviation provides an estimate of 
how much the prices have deviated from the average in the past. Higher standard deviation 
indicates higher volatility.

Volatility Index (VIX): The VIX is a popular index that measures the implied volatility of the options market.
It reflects the market's expectations of future volatility. The VIX is often referred to as the "fear index" as it tends to rise during times of market stress and uncertainty.
Volatility plays a crucial role in various aspects of finance, such as risk management, option pricing, portfolio construction, and trading strategies. It helps investors and traders assess the potential risks and rewards associated with different investments and make informed decisions.

It's important to note that volatility is not the same as the direction of price movements. 
A highly volatile asset can have significant price swings in both positive and negative directions, 
while a less volatile asset may have relatively smaller and more predictable price movements.

By understanding and analyzing volatility, market participants can better evaluate the potential 
risks and returns of their investments and adapt their strategies accordingly.

### Rolling

Concept of "rolling" refers to a technique of calculating metrics or performing calculations repeatedly on a moving or rolling window of data. Rolling calculations are commonly used to analyze time-series data, such as stock prices, economic indicators, or financial ratios, in order to observe trends or patterns over a specified period.

The rolling window typically consists of a fixed number of consecutive observations or a defined time period. As new data becomes available, the window moves forward in time, and the calculations are updated using the latest observations while dropping the oldest ones.

With rolling  we can have rolling mean ( Moving averages), Rolling standard deviation (Volality), rolling sum, rolling corelations

### Techinical analysis
On-Balance Volume (OBV): is a technical analysis indicator used in finance to measure the cumulative 
buying and selling pressure of a security over a specified period. 
It is based on the theory that changes in volume can precede price movements.
OBV is calculated by adding the volume of an asset to a running total if the price of the asset closes higher than the previous day's closing price. If the price closes lower than the previous day's closing price, the volume is subtracted from the running total. If the price remains unchanged, the volume is not added or subtracted. 
The resulting OBV line can help identify trends and potential reversals in the price of the security.
    On-balance Volume (OBV)

Where

    OBV: On-balance Volume level
    OBV_{Previous}: Previous On-balance Volume level
    volume: Last trending volume amount

```python
%%timeit
def cal_obv(df):
    df = df.copy()
    df['OBV'] = 0.0
    for i in range(len(df)):
        if df['Close'][i] > df['Close'][i-1]:
            df['OBV'][i] = df['OBV'][i-1]+ df['Volume'][i]
        if df['Close'][i] < df['Close'][i-1]:  
            df['OBV'][i] = df['OBV'][i-1]- df['Volume'][i]
        if df['Close'][i] == df['Close'][i-1]: 
            df['OBV'][i] = df['OBV'][i-1]
    return df        
    
## Loop 

#cal_obv(app)
```

### Acumulation/ Distribution Indicator (A/D) :
The Accumulation/Distribution Indicator (A/D) is a technical analysis tool used to assess the flow of money 
into or out of a security. It is based on the premise that the volume of trading activity is an important 
factor in determining the strength and direction of a price trend.

The A/D indicator calculates a cumulative value by considering both the price and volume data. 
It is typically used to confirm the strength of a trend and identify potential reversals.

The formula for calculating the Accumulation/Distribution Indicator (A/D) is as follows:

A/D = Previous A/D + CMFV
where CMFV is current money flow Volume

MFV= MFM  * Volume    ,MFV is money flow Volume

MFM =[(Close - Low) - (High - Close)] / (High - Low) ,MFM is  Money Flow Multiplier

Where:

Previous A/D: The value of the A/D indicator from the previous period.
Close: The closing price of the current period.
Low: The lowest price of the current period.
High: The highest price of the current period.
Volume: The volume of trading activity during the current period.

The A/D indicator takes into account whether the closing price is closer to the high or low of the period, 
as well as the volume traded. If the closing price is closer to the high, it suggests buying pressure, 
while if it is closer to the low, it suggests selling pressure.
The A/D line can be plotted on a chart along with the price of the security. 
When the A/D line is moving in the same direction as the price, it confirms the strength of the trend. 
Divergence between the A/D line and the price can indicate potential trend reversals.

Traders and analysts use the A/D indicator to gauge the demand and supply dynamics of a security 
and make informed decisions about buying or selling. It can help identify potential buying or selling opportunities and confirm the sustainability of a price trend.

### Relative strength Index
The Relative Strength Index (RSI) is a popular technical analysis indicator used to measure the speed and change of price movements in a security. It helps traders and analysts identify overbought and oversold conditions, potential trend reversals, and the overall strength of a price trend.

The RSI is calculated based on the average gains and losses over a specified period, typically 14 days. The formula for calculating the RSI is as follows:

RSI = 100 - (100 / (1 + RS))

Where:

RS (Relative Strength) = Average Gain / Average Loss

Average Gain = Average of the gains over the specified period
Average Loss = Average of the losses over the specified period

The RSI value ranges from 0 to 100. Traditionally, an RSI value above 70 is considered overbought, indicating a potential price reversal or correction to the downside. Conversely, an RSI value below 30 is considered oversold, suggesting a potential price reversal or bounce to the upside. These threshold values can be adjusted based on the specific security or trading strategy.

The RSI is typically displayed as a line chart with values ranging from 0 to 100. Traders often look for bullish or bearish divergences between the RSI and the price, where the RSI moves in the opposite direction to the price, signaling a potential change in trend.

It's important to note that the RSI is just one tool among many used in technical analysis, and it should be used in conjunction with other indicators and analysis techniques to make well-informed trading decisions. The RSI is most effective when combined with other indicators or used in conjunction with support and resistance levels, trendlines, or other chart patterns.
